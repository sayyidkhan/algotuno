```javascript
var csv = require('csv-parser');
var fs = require('fs');
```

```javascript
csv_result = []

fs.createReadStream('./dataset/gold_price.csv')
  .pipe(csv())
  .on('data', (row) => {
    csv_result.push(row);
    console.log(row);
  })
  .on('end', () => {
    console.log('CSV file successfully processed');
});
```

```javascript
var updated_csv_result = csv_result.map(e => {
   var price = e['USDI_Price'];
   var date = e['Date'];
   var obj = {
       'usd_price' : price,
       'date' : new Date(date),
   }
   return obj;
});
updated_csv_result;
```

```javascript
var sort_updated_csv_result = updated_csv_result.sort(function(a,b){
  // Turn your strings into dates, and then subtract them
  // to get a value that is either negative, positive, or zero.
  return new Date(b.date) - new Date(a.date);
});
sort_updated_csv_result;
```

```javascript
// write into filesystem
var jsonData = JSON.stringify(sort_updated_csv_result);
fs.writeFile("./dataset/gold_price.json", jsonData,'utf8', function(err) {
    if (err) {
        console.log(err);
    }
});
```
